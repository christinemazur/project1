```python
# Dependencies and Setup
import pandas as pd

# File to Load
NYC1_data_to_load = "resources/NYC_Listings1.csv"
NYC2_data_to_load = "resources/NYC_Listings2.csv"
NYC3_data_to_load = "resources/NYC_Listings3.csv"

SF_data_to_load = "resources/SF_Listings.csv"

# Read School and Student Data File and store into Pandas Data Frames
NYC1_data = pd.read_csv(NYC1_data_to_load)
NYC2_data = pd.read_csv(NYC2_data_to_load)
NYC3_data = pd.read_csv(NYC3_data_to_load)


SF_data = pd.read_csv(SF_data_to_load)

# Combine the data into a single dataset
##airbnb_data_complete = pd.merge(NYC_data, SF_data, how="leftORright", on=["column1", "column2"])
```

```python
# Create a full NYC DF
NYC_data = pd.concat([NYC1_data, NYC2_data, NYC3_data], ignore_index=True)
# ADD state columns to NYC and SF data
NYC_data['state'] = 'NY'
SF_data['state'] = 'SF'

# Concatenate NYC an
airbnb_data = pd.concat([NYC_data, SF_data], ignore_index=True)
airbnb_data.groupby('state').mean()
```
